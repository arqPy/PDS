```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gerar um sinal
t = np.arange(-1, 3.7,0.1) # Base de tempo - Variável Independente

y = 2*t - 4 # Sinal: Variável Dependente de t

plt.plot(t, y) # Usado para função contínua

plt.grid('on')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')
plt.title('$y = 2t - 4$')
```

```python
# Gerar um sinal
#n = np.arange(-1, 10,1) # Base de tempo - Variável Independente
n = np.arange(-1, 10, 1)

y = 2*n - 4 # Sinal: Variável Dependente de t

plt.grid('y')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')
plt.title('$y = 2n - 4$')
plt.ylim(-6, 15)
plt.stem(n, y) # Usado para função discreta

```

```python
def degrauC(t):
  u = (t >= 0) * 1.                     # programação funcional
  return u

def degrauD(n):
  u = (n >= 0) * 1.
  return u

n = np.arange(-1, 2, .1)
t = np.arange(-1, 2, .1)

ya = np.sin(2*np.pi*t)
yd = np.sin(2*np.pi*n)

plt.grid()
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')

plt.plot(t, ya, 'green')
plt.stem(n, yd, 'blue')
```

```python
plt.plot(t, degrauC(t), 'green')
plt.stem(n, degrauD(n), 'blue')
```
