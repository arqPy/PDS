```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gerar um sinal
t = np.arange(-1, 3.7,0.1) # Base de tempo - Variável Independente

y = 2*t - 4 # Sinal: Variável Dependente de t

plt.plot(t, y) # Usado para função contínua

plt.grid('on')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')
plt.title('$y = 2t - 4$')
```

```python
# Gerar um sinal
#n = np.arange(-1, 10,1) # Base de tempo - Variável Independente
n = np.arange(-1, 10, 1)

y = 2*n - 4 # Sinal: Variável Dependente de t

plt.grid('y')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')
plt.title('$y = 2n - 4$')
plt.ylim(-6, 15)
plt.stem(n, y) # Usado para função discreta

```

```python
def degrauC(t):
  u = (t >= 0) * 1.                     # programação funcional
  return u

def degrauD(n):
  u = (n >= 0) * 1.
  return u

n = np.arange(-1, 2, .1)
t = np.arange(-1, 2, .1)

ya = np.sin(2*np.pi*t)
yd = np.sin(2*np.pi*n)

plt.grid()
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude (V)')

plt.plot(t, ya, 'green')
plt.stem(n, yd, 'blue')
```

```python
plt.plot(t, degrauC(t), 'green')
plt.stem(n, degrauD(n), 'blue')
```

```python
def degrauD(n, n0=0):
  u = (n >= n0) * 1.
  return u
```

```python
n = np.linspace(-1, 10, 20) #.arange(-1, 12, .5) # significa que Ts eh igual a .5
plt.stem(n, degrauD(n, 4), 'blue')
```

```python
t = np.arange(-2,9)
t1 = t + 2 # deslocamento
t2 = -t # inversão
x = [ 0, 0, 0, 1, 2, 3, 2, 1, 0, 0, 0]
plt.stem(t,x, 'green')
#plt.stem(t1,x, 'blue')
plt.stem(t2, x, 'red')
plt.show()
'''
t = np.arange(-2,9) + 2 # deslocamento de 2 unidades, manter a dimensçao de x
plt.stem(t,x)
plt.show()
'''
```

```python
# Escalamento (ou Mudança de Escala) de Tempo
# Down-sampling
x1=x[0:9:2]
t1 = np.arange(-1,4)
plt.stem(t1,x1)
plt.show()
```

```python
# Up-sampling
x1 = np.zeros(16)
for t in range(0,15):
  if (t % 2 == 0):
    x1[t] = x[t//2]
  else:
    x1[t] = 0
t1 = np.arange(-4,12)
plt.stem(t1,x1)
plt.show()
```

# Modelos Importantes de Sinais em Tempo Discreto:


*  Impulso Unitário
*  Rampa




```python
n = np.arange(-7,7)
k = 3 # Instante do Impulso
y = (n == k) * 1
plt.stem(n,y)
plt.grid()
```

```python
y = 2*n
plt.stem(n,y)

plt.grid('on')
```

```python
larg = 15;                             # largura da janela (somente valor ímpar)
mtd = (larg-1)/2                       # metade da largura
n = np.arange(-20,20)                     # base de tempo
w = np.zeros(len(n))                      # inicialização com zeros
w[(n >= -mtd)*(n <= mtd)] = 1.         # amostras unitárias
plt.stem(n,w)
plt.grid(True)
```
